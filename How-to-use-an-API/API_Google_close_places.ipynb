B. Create a function to Find the Place Id given Place Name (String) and Location (Lat/Lont Tuple).

```python
def find_place(key, input1, inputtype, locationbias = None):
    '''
    '''
    
    url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json"
    
    parameters = {'key':key,'input':input1,'inputtype':inputtype,'locationbias':'locationbias =' + str(locationbias)}
    res = requests.get(url, params = parameters)
    results = json.loads(res.content.decode('utf-8'))
    
    return results['candidates'][0]['place_id']
```

D. Use the Place Details Method to find the Reviews of a Place in English and Spanish. In order to fulfill this item you need to:

    Create a find_reviews function with the appropiate parameters. It must return a JSON with the results.
    Create a table with the following columns: Author Name, Language, Rating, Text, Time.



```python
def find_reviews(place_id,key):
    '''
    
    find_reviews is a funtion which will make the first request to get the reviews data in json format.
    
    Args:
    
        place_id: It depends on the desired place that we want to look for.
        
        key: our Api-key.
        
    Returns:
    
        results : This variable will contain a dictionary with the previously defined data.
    
    '''
    
    url = 'https://maps.googleapis.com/maps/api/place/details/json?place_id={}&language=en,es&key={}'.format(place_id,key)
    res = requests.get(url)
    resultados = json.loads(res.content)
    
    return resultados
```

E. Find similar places to Los Trompos Circuito. Rank them by its "Prominence". In order to fulfill this item you need to:

    Create a find_similar function with the appropiate parameters. It must return a JSON with the results.
    Create a table with the following columns: Place Name, Plaece Id, Rating, Total of Use Ratings.



```python
def find_similar(name, latitude, longitude,key):
    '''
    
    This function is going to analize and give us the nerby places in a defined place.
    
    Parameters:
            name:      which contains what restaurant you are looking for.
            latitude:  defined area.
            longitude: defined area.
            key:       Our Apy to get the access.
    
    
    Returns: 
    
            results: which contains a dictionary with all desired data.
            
    '''
    #
    #First of all, we need to take the first places (which are on the top.)
    url = f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={name}&inputtype=textquery&fields=types&key={key}'
    res = requests.get(url)
    results = json.loads(res.content)
    place_type = results['candidates'][0]['types'][0]
    
    #now we use the previous function to get the nearby places.
    url_2 = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={latitude},{longitude}&radius=2000&rankby=prominence&type={place_type}&key={key}'
    res_2 = requests.get(url_2)
    results_nearby = json.loads(res_2.content)
    
    return results_nearby['results']
```

```python
# we use the function and we save the result in a variable
similar_data = find_similar("Los Trompos Circuito",20.988459,-89.736768,key)

# Let´s read the Data Frame to start working with the data.
find_similar = pd.read_json(json.dumps(similar_data ), orient='records')[['name','place_id','rating','user_ratings_total']]

#Renme the columns.
cols_name = {'name':'Place Name','place_id':'Plaece Id','rating': 'Rating','user_ratings_total':'Total of Use Ratings'}
find_similar = find_similar.rename(columns = cols_name)

find_similar.head()
```
